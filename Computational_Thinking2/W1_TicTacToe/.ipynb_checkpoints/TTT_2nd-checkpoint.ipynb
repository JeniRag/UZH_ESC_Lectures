```python
"""Board """
class game():
    def __init__(self):
        self.field=[None, None, None, None, None, None, None, None, None]
        self.empty_field=9
        self.path_order=[]
        self.user_order=[]
        
    def is_valid(self, n):
        if n<0 or n>=9:
            print("Not existing Field ")
            return False
        
        elif self.field[n] !=None:
            print("Already occupied")
            return False
        
        else:
            #print("valid")
            return True
        
    def setX(self, n, option=True): #User
        if (self.empty_field !=0 and self.is_valid(n)==True):
            self.field[n]="X"
            self.path_order.append(n)
            self.user_order.append(1)
            if option==True:
                self.empty_field-=1
    
    def setO(self, n, option=True): #Computer
        if (self.empty_field !=0 and self.is_valid(n)==True):
            self.field[n]="O"
            
            self.path_order.append(n)
            self.user_order.append(2)
            if option==True:
                self.empty_field -=1
            
    def ask_pos(self):
        pos=input("Which field?: ")
        pos=int(pos)
        while self.is_valid(pos)==False:
            self.ask_pos()
        return pos
    
    def show(self):
        print(self.field[0:3])
        print(self.field[3:6])
        print(self.field[6:])
        print(" ")
    
    def is_ended(self, option=False):
        
        if self.win(1)==True:
            if option==True:
                print("X wins")
            return True
        
        elif self.win(2)==True:
            if option==True:
                print("O wins")
            return True
        
        elif self.field.count(None)==0:#count number of elements
            if option==True:
                print("Unentschieden")
            return True
        
        else:
            return False
    def row_win(self,player):
        if player==1:
            sign="X"
        else:
            sign="O"
        li=[]
        
        for i in range(0,6,3):# 0, 3, 6
            if (self.field[i]== sign and self.field[i+1]==sign and self.field[i+2]==sign):
                li.append(True)
            else:
                li.append(False)
        return any(li)==True
            
    def col_win(self, player):
        if player==1:
            sign="X"
        else:
            sign="O"
        li=[]
        for i in range(0,3): #0,1,2
            if self.field[i]== sign and self.field[i+3]==sign and self.field[i+6]==sign:
                li.append(True)
            else:
                li.append(False)
        return any(li)==True
            
    def diag_win(self,player):
        if player==1:
            sign="X"
        else:
            sign="O"
        if (self.field[0]==sign and self.field[4]==sign and self.field[8]==sign) or (self.field[2]==sign and self.field[4]==sign and self.field[6]==sign):
        
            return True
        else:
            return False
        
    def win(self,player):
        if (self.diag_win(player)==True or self.row_win(player)==True or self.col_win(player)==True):
            return True
        else:
            return False
    
    def empty(self):
        empty_list=[]
        for i in range(len(self.field)):
            if self.field[i]==None:
                empty_list.append(i)
        return empty_list
                
           
```

```python
class node(): #possible moves for tree
    def __init__(self):
        self.score=None # -1, 0, 1
        self.pos=None #position    
```

```python
def points(g2):

    if (g2.win(1)==True):
      #if first player wins
        #g2.show()
        return 1
        
    elif (g2.win(2)==True) : #Computer will minimieren
        #g2.show()
        return -1
    
    elif g2.field.count(None)==0: #Unentschieden
        #g2.show()
        return 0
    
    #else:
       
     #   return False #not terminal state
    
def minmax(g2, player):
    
    if g2.is_ended()==True: #if in terminal state
        res= points(g2)
        return res
    

    else :
        if player==1:
            #player maximize
            best=-2
            for i in range(9):
                if g2.field[i]==None: # if cell empty
                    g2.field[i]= "X" #fill with character
                    best=max(best, minmax(g2, 2)) #minmax of opponent
                    g2.field[i]=None #return to normal
            #print("resulutat: "+ str(best))
            return best


        else: #player 2 minimize
            best=2
            for i in range(9):
                if g2.field[i]==None:
                    g2.field[i]= "O" #Computer
                    best=min(best, minmax(g2,1)) 
                    g2.field[i]=None
            #print("resulutat: "+ str(best))
            return best
        
def decide_move(g2, player):

    n1=node()
    
    #n1.pos=0
    max_val=-2
    min_val=2
     
    if player==1:
        for i in range(9):
            if g2.field[i]==None:
            
                g2.field[i]="X"
                res=minmax(g2,2) ###number correct????????
                #print("result: "+ str(res)) 
                if max_val<res:
                    max_val=res
                    n1.pos=i
                    
                    
                g2.field[i]=None   
                
    else:  #player 2's turn
        for i in range(9):
            if g2.field[i]==None:
                g2.field[i]="O" #Computer
                res=minmax(g2,1)
                
                if res<min_val:
                    min_val=res
                    n1.pos=i 
                    
                g2.field[i]=None
            
            
            
            
                
    #print("best move: " + str(n1.pos))
    return n1.pos
```
